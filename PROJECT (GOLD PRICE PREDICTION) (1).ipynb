```python
# Importing libraries
import numpy as np 
import matplotlib.pyplot as plt  
import pandas as pd 
```

```python
# Importing datasets
data_set=pd.read_csv("FINAL_USO.csv")
```

```python
data_set
```

```python
t=data_set.drop(["Date"],axis=1)
```

```python
# Extracting independent and dependent variable 
x= t.iloc[:, :-79].values  
y= data_set.iloc[:, 5].values
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 1/3, random_state=0) 
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Fitting the Simple Linear Regression model to the training dataset 
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(x_train, y_train)
```

```python
# Prediction of Test and Training set result
y_pred= regressor.predict(x_test)  
x_pred= regressor.predict(x_train)
```

```python
y_pred
```

```python
x_pred
```

```python
# Visualizing the training set results 
plt.scatter(x_train, y_train, color="green")   
plt.plot(x_train, x_pred, color="red")    
plt.title("Open v/s Adj close")  
plt.xlabel("Open")  
plt.ylabel("Adj close")  
plt.show()
```

```python
# Visualizing the Test set results
plt.scatter(x_test, y_test, color="blue")   
plt.plot(x_train, x_pred, color="red")    
plt.title("Open v/s Adj close")  
plt.xlabel("Open")  
plt.ylabel("Adj close")  
plt.show()
```

```python
print('Train Score: ', regressor.score(x_train, y_train))  
print('Test Score: ', regressor.score(x_test, y_test))
```

```python
# Importing libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importing datasets
t=data_set.drop(["Date"],axis=1) 
```

```python
# Extracting Independent and dependent Variable 
x= t.iloc[:, :-79].values  
y= data_set.iloc[:, 5].values
```

```python
x
```

```python
y
```

```python
# Fitting the Linear Regression to the dataset
from sklearn.linear_model import LinearRegression  
lin_R= LinearRegression()  
lin_R.fit(x,y)
```

```python
# Fitting the Polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures  
poly_R= PolynomialFeatures(degree= 5)   # Change the degree for the correct output
x_poly= poly_R.fit_transform(x)  
lin_R2 =LinearRegression()  
lin_R2.fit(x_poly, y)
```

```python
# Visulaizing the result for Linear Regression model 
plt.scatter(x,y,color="blue")  
plt.plot(x,lin_R.predict(x), color="red")  
plt.title("Open v/s Adj close")  
plt.xlabel("Open")  
plt.ylabel("Adj close")  
plt.show()
```

```python
# Visulaizing the result for Polynomial Regression 
plt.scatter(x,y,color="blue")  
plt.plot(x, lin_R2.predict(poly_R.fit_transform(x)), color="red")  
plt.title("Open v/s Adj close (Polynomial Regression)")  
plt.xlabel("Open")  
plt.ylabel("Adj close")  
plt.show()
```

```python
lin_pred = lin_R.predict([[156.983456]])  
print(lin_pred)
```

```python
poly_pred = lin_R2.predict(poly_R.fit_transform([[156.983456]]))  
print(poly_pred)
```

```python
Multilinear Regression method has more accuracy.
```
